# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

# Problem 1

```python
##Define the x and k values used to generate the Gaussian and the Fourier transform
x = np.arange(-10, 10, 0.1)
k = np.arange(x.size)

#Make the Gaussian function
def gauss(x, mu, sig):
    return np.exp(-(x-mu)**2/(2*sig**2))/np.sqrt(2*np.pi*sig**2)

#Get the gaussian that we are going to shift. Center it at 0 and give a standard deviation of 1.5.
y = gauss(x, 0, 1.5)

#Define the function to shift your array
def shift(ar, shift):
    #Get the fourier transform for the intial gaussian
    ft = np.fft.fft(ar)
    #Get the value of N, the array size in our case
    N = ar.size
    #Get the exponential that we use to shift our function
    sh = np.exp((-2*np.pi*1J*shift*k)/N)
    #Get the shifted gaussian
    y = np.fft.ifft(sh*ft)
    return np.real(y)

#Plotting the original gaussian and the one shifted by half the array length
newy = shift(y, x.size/2)
plt.plot(x, newy, 'r', label='Shifted Gaussian')
plt.plot(x, y, 'b', label = 'Non-shifted Gaussian')
plt.title('Shifting an array with a convolution')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('./gauss_shift.png')
plt.show()
```

# Problem 2

```python
#Making the correlation function, using the fact that f*g = ift(dft(f) x conj(dft(g)))
def correlation(ar1, ar2):
    #Getting the fourier transform of f
    y1 = np.fft.fft(ar1)
    #Getting the conjugate of the fourier transform of f
    y2 = np.conjugate(np.fft.fft(ar2))
    return np.real(np.fft.ifft(y1*y2))

#Getting the gaussian function that we will correlate with itself. 
gauss1 = gauss(x, 0, 3)

#Calculate and plot the correlation of the gaussian with itself.
gausscorr = correlation(gauss1, gauss1)

plt.plot(x, gauss1, 'r', label='Gaussian function')
plt.plot(x, gausscorr, 'b', label='Correlation of gaussian')
plt.title('Correlation of a Gaussian with itself')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('./Prob2_plot.png')
plt.show()
```

# Problem 3

```python
#Creating two gaussians, where the second one corresponds to the first one shifted.
gauss1 = gauss(x, 0, 1.5)
gauss2 = shift(gauss1, 250)
#Get the correlation of the shifted and non-shifted gaussians using the correlation function
#defined in Problem 2.
gausscorr = correlation(gauss1, gauss2)
#Plotting the result of the correlation.
plt.plot(x, gausscorr, 'b', label='Corellated Gaussian')
plt.plot(x, gauss2, 'r', label='Shifted Gaussian')
plt.plot(x, gauss1, 'k', label='Non-shifted Gaussian')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('./Prob3_plot.png')
plt.show()
```

# Problem 4

```python
#Defining 3 different x arrays of different lengths to test out if 
#our function does indeed account for the difference in lengths.
#We chose to use initial arrays of length 100 and 200 for the gaussians. 
#Since our function doubles the size of the output we also had to define
#a third x array for the output of the safe convoluton function.
x = np.linspace(-10, 10, 100)
newx = np.linspace(-10, 10, 200)
newnewx = np.linspace(-10, 10, 400)

#Getting the two gaussians that we will convolve.
gauss1 = gauss(x, 0, 1.5)
gauss2 = gauss(newx, 9, 1.5)

#Making a safe convolution function.
def conv_safe(f, g):
    #Separating the different cases, where f and g have different lengths.
    #We double the size of the arrays anf fill them with zeroes initially.
    if f.size < g.size:
        new_f = np.zeros(2*g.size)
        new_g = np.zeros(2*g.size)
    elif g.size < f.size:
        new_f = np.zeros(2*f.size)
        new_g = np.zeros(2*f.size)
    else:
        new_f = np.zeros(2*f.size)
        new_g = np.zeros(2*g.size)
        
    #After doubling the size of our input arrays, we fill them with the 
    #f and g values at the beginning and leave the ends filled wth zeroes. 
    new_f[:f.size]=f
    new_g[:g.size]=g
    #We calculate the convolution using f and g's Fourier transforms.
    ff1 = np.fft.fft(new_f)
    ff2 = np.fft.fft(new_g)
    return np.real(np.fft.ifft(ff1*ff2))

#Get the safe convolution of f and g.
gauss3 = conv_safe(gauss1, gauss2)

#Plot the 2 inital functions and their non-circulant convolution.
plt.plot(x, gauss1, 'b', label='First Gaussian')
plt.plot(newx, gauss2, 'r', label='Second Gaussian')
plt.plot(newnewx, gauss3, 'g', label='Safe Convolution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the Save Convolution method')
plt.legend()
plt.savefig('./Prob4_plot.png')
plt.show()
```

# Problem 5b - Second Part

```python
#Make the function to calculate the fraction
def frac(k, N):
    num = 1-np.exp(-2*np.pi*1J*k)
    den = 1-np.exp((-2*np.pi*1J*k)/N)
    return num/den
#Make a list of k values and select a value for N
Klist = np.linspace(1, 100,100)
N = 10
#Get the fraction's value at each K value
myy=[]
for k in Klist:
    myy.append(frac(k, N))
    
#Plot the values of the fraction for different values of k
#As we can see, the fraction is equal to 0 everywhere except when k is a multiple of N.
plt.plot(Klist, np.abs(myy), '.')
plt.xlabel('k')
plt.ylabel('Value of the sum')
plt.title('Plot of the value of the sum against different values of k')
plt.savefig('./Prob5b2_plot.png')
print('As we can see, when k is a multiple of N, the sum returns a value, but ' 
      'when k is not a multiple of N, the sum is equal to zero.')
```

# Problem 5c

```python
#See written part for proof that sin function can be written in this form.
#We use this function as our analytical Fourier transform of the non-integer period sine function.
def mysin(k, kprime, N):
    A = (1-np.exp(2*np.pi*1J*(k-kprime)))/(1-np.exp(2*np.pi*1J*(k-kprime)/N))
    B = (1-np.exp(-2*np.pi*1J*(k+kprime)))/(1-np.exp(-2*np.pi*1J*(k+kprime)/N))
    return np.abs(((A-B)/(2J)))
    
#Make a function for the numerical Fourier transform of the sine function.
def numsin(a, x, N):
    A = np.fft.fft(np.sin(2*np.pi*a*x/N))
    return np.abs(A)

#Define our values for N and the non-integer a.
N=50
a = 2.7

#Define our x array for real space and k array for Fourier space.
x = np.linspace(1, N, N)
k = np.fft.fftfreq(x.size, 1/N)

#Get the analytical and numerical Fourier transforms for given values of x, k, a and N.
my_y = mysin(a, k, N)
num_y = numsin(a, x, N)

#Plot the Analytical and Numerical Fourier transforms
plt.plot(x, my_y, 'b', label='Analytic DFT')
plt.plot(x, num_y, 'r', label='Numerical DFT')
plt.xlabel('k')
plt.ylabel('DFT')
plt.title('Plot of the analytic and numerical DFT of a sine wave, with N=50')
plt.legend()
plt.savefig('./Prob5c_plot_50.png')
plt.show()

#Calculating and plotting the residuals, to see if we have achieved machine epsilon order magnitude.
residual = np.abs(my_y - num_y)
plt.plot(x, residual)
plt.xlabel('k')
plt.ylabel('DFT')
plt.title('Residuals')
plt.savefig('./Prob5c_residuals_plot.png')
plt.show()

print('As we can see in the Residuals, the FFT agrees with our analytical solution, but not within machine '
      'epsilon. I believe there has to be some error in my way of defining the analytical solution that accounts '
       'for this.')
print('The Fourier transform of my analytical solution looks vaguely like a delta function, but if we reduce '
      'N we can see that this is not the case.')
```

# Problem 5d

```python
#Defining the window function
def window(x, N):
    return 0.5 - 0.5*np.cos((2*np.pi*x)/N)

#Convolving the window function with my analytical solution.
window_my_y = np.abs(np.fft.fft(np.fft.ifft(my_y)*window(x, N)))

plt.plot(x, window_my_y, 'b', label='Analytic DFT')
plt.plot(x, num_y, 'r', label='Numerical DFT')
plt.xlabel('k')
plt.ylabel('DFT')
plt.legend()
plt.title('Plot of the analytic and numerical DFT of a sine wave, with a window function')
plt.savefig('./Prob5d_plot.png')
plt.show()

window_residuals = np.abs(window_my_y - num_y)
plt.plot(x, window_residuals)
plt.xlabel('k')
plt.ylabel('DFT')
plt.title('Residuals with a window function')
plt.savefig('./Prob5d_residuals_plot.png')
plt.show()

print('As we can see, my analytical solution did not improve with the use of a window function. '
      'I suspesct this is due to my earlier error in Problem 5c that is reverberating in the resuts here.')
```
